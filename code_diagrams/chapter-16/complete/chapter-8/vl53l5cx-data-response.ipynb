```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
distance_mm = np.arange(0, 64)
```

```python
# Reshape the array into an 8 by 8 grid
distance_mm = distance_mm.reshape((8, 8))
distance_mm = np.fliplr(distance_mm)

print(distance_mm)
```

```python
fig, ax = plt.subplots()
# Create a grid diagram all white
ax.matshow(np.ones((8, 8)), cmap='gray_r')
# Add grid lines
ax.grid(True)

# Adjust the limits of the x and y axes
ax.set_xlim(0, 8)
ax.set_ylim(0, 8)

# Annotate each cell with the numeric value
for i in range(distance_mm.shape[0]):
    for j in range(distance_mm.shape[1]):
        ax.text(j + 0.5, i + 0.5, str(distance_mm[i, j]), va='center', ha='center', color='black')
ax.xaxis.set_ticks_position('bottom')

# Add labels below the x axis, with "robots left" on bottom left and "robots right" on bottom right
ax.text(1, -1, "Robot's bottom left", va='center', ha='center', color='black')
ax.text(6.5, -1, "Robot's bottom right", va='center', ha='center', color='black')

plt.show()
```
