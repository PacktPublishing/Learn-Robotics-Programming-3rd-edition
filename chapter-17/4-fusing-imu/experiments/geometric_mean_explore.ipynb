```python
import numpy as np
import matplotlib.pyplot as plt
```

# Geometric Mean vs Iteration
Use a short sequence of positive values, then compute the cumulative geometric mean after each new value.

```python
values = np.array([0.9, 0.7, 0.8, 0.01, 0.95, 0.6, 0.85, 0.5], dtype=float)
iterations = np.arange(1, len(values) + 1)

log_values = np.log(values)
cumulative_geom_mean = np.exp(np.cumsum(log_values) / iterations)

for i, gm in zip(iterations, cumulative_geom_mean):
    print(f"Iteration {i}: geometric mean = {gm:.4f}")
```

```python
plt.figure(figsize=(8, 4))
plt.plot(iterations, cumulative_geom_mean, marker="o", linewidth=2)
plt.xticks(iterations)
plt.xlabel("Iteration")
plt.ylabel("Cumulative geometric mean")
plt.title("Cumulative Geometric Mean Across Iterations")
plt.grid(True, alpha=0.3)
plt.tight_layout()
```

```python
# Can we plot the product of the values across iterations?
cumulative_product = np.cumprod(values)
plt.figure(figsize=(8, 4))
plt.plot(iterations, cumulative_product, marker="o", linewidth=2)
plt.xticks(iterations)
plt.xlabel("Iteration")
plt.ylabel("Cumulative product")
plt.title("Cumulative Product Across Iterations")
plt.grid(True, alpha=0.3)
plt.tight_layout()
```

```python
cumulative_mean = np.cumsum(values) / iterations

plt.figure(figsize=(8, 4))
plt.plot(iterations, cumulative_mean, marker="o", linewidth=2)
plt.xticks(iterations)
plt.xlabel("Iteration")
plt.ylabel("Cumulative mean")
plt.title("Cumulative Mean Across Iterations")
plt.grid(True, alpha=0.3)
plt.tight_layout()
```

```python
fig, axes = plt.subplots(3, 1, figsize=(8, 10))

axes[0].plot(iterations, cumulative_geom_mean, marker="o", linewidth=2)
axes[0].set_xticks(iterations)
axes[0].set_ylabel("Cumulative geometric mean")
axes[0].set_title("Cumulative Geometric Mean Across Iterations")
axes[0].grid(True, alpha=0.3)

axes[1].plot(iterations, cumulative_product, marker="o", linewidth=2)
axes[1].set_xticks(iterations)
axes[1].set_ylabel("Cumulative product")
axes[1].set_title("Cumulative Product Across Iterations")
axes[1].grid(True, alpha=0.3)

axes[2].plot(iterations, cumulative_mean, marker="o", linewidth=2)
axes[2].set_xticks(iterations)
axes[2].set_xlabel("Iteration")
axes[2].set_ylabel("Cumulative mean")
axes[2].set_title("Cumulative Mean Across Iterations")
axes[2].grid(True, alpha=0.3)

plt.tight_layout()
```
