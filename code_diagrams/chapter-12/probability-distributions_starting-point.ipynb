```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
uniform_dist = np.random.uniform(-4, 4, 10000)
normal_dist = np.random.normal(0, 1, 10000)

plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.hist(uniform_dist, bins=30, density=True, alpha=0.6, color='#777777')
plt.ylim(0, 0.5)
plt.xlim(-4, 4)
plt.title('Uniform Distribution', fontsize=20)
plt.xticks(ticks=plt.xticks()[0], labels=[])  # Keep ticks but hide x-axis labels
plt.yticks(ticks=plt.yticks()[0], labels=[])  # Keep ticks but hide y-axis labels

plt.subplot(1, 2, 2)
plt.hist(normal_dist, bins=30, density=True, alpha=0.6, color='#777777')
plt.ylim(0, 0.5)
plt.xlim(-4, 4)
plt.title('Uniform Distribution', fontsize=20)
plt.xticks(ticks=plt.xticks()[0], labels=[])  # Keep ticks but hide x-axis labels
plt.yticks(ticks=plt.yticks()[0], labels=[])  # Keep ticks but hide y-axis labels
plt.tight_layout()
```
