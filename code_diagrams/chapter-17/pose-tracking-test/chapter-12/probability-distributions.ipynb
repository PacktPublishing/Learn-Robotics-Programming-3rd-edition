```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
uniform_dist = np.random.uniform(-4, 4, 10000)
normal_dist = np.random.normal(0, 1, 10000)

plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.hist(uniform_dist, bins=30, density=True, alpha=0.6, color='#777777')
plt.ylim(0, 0.5)
plt.xlim(-4, 4)
plt.title('Uniform Distribution', fontsize=20)
plt.xticks(ticks=plt.xticks()[0], labels=[])  # Keep ticks but hide x-axis labels
# Hide borders
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['left'].set_visible(False)
# Hide y-axis
plt.gca().yaxis.set_visible(False)

plt.subplot(1, 2, 2)
plt.hist(normal_dist, bins=30, density=True, alpha=0.6, color='#777777')
plt.ylim(0, 0.5)
plt.xlim(-4, 4)
plt.title('Normal Distribution', fontsize=20)
plt.xticks(ticks=plt.xticks()[0], labels=[])  # Keep ticks but hide x-axis labels
# Hide borders
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['left'].set_visible(False)
# Hide y-axis
plt.gca().yaxis.set_visible(False)
# Set a higgher resolution for the layout
plt.gcf().set_dpi(200)
plt.tight_layout()
```
