```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
images_to_test = [
    "line_1.jpg",
    "line_2.jpg",
    "line_3.jpg",
    "line_4.jpg",
    "line_5.jpg",
    "line_6.jpg",
    "carpet_line1.jpg",
    "empty_carpet.jpg",
    "empty_gray.jpg"
]
```

```python
for image in images_to_test:
    print("Testing image:", image)
    image = cv2.imread(image)
    resized = cv2.resize(image, (320, 240))
    gray = cv2.cvtColor(resized, cv2.COLOR_BGR2GRAY)
    plt.figure(figsize=(10, 5))
    plt.subplot(1, 5, 1)
    plt.imshow(gray, cmap='gray')
    plt.subplot(1, 5, 2)
    masked = cv2.inRange(gray, 0, 50)
    plt.imshow(masked, cmap='gray') 
    plt.subplot(1, 5, 3)
    _, masked = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
    plt.imshow(masked, cmap='gray') 
    plt.subplot(1, 5, 4)
    block_size = 65
    mean_subtracted_constant = 10
    masked = cv2.adaptiveThreshold(gray, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, block_size, mean_subtracted_constant)
    plt.imshow(masked, cmap='gray') 
    plt.subplot(1, 5, 5)
    masked = cv2.adaptiveThreshold(gray, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, block_size, mean_subtracted_constant)
    plt.imshow(masked, cmap='gray') 
    plt.show()
```

It's pretty clear that the simple 50 threshold is giving the most usable results. Keep it simple is robust against any more complex thresholding.
